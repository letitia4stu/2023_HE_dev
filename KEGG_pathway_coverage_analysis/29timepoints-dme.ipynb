```python
dfs=pd.read_csv(r'/Users/lichaoran/Downloads/figure2/pathplot/slide-Notch-path-zhanshi-exp-deduplicated.csv',index_col=0)
dfs
```

```python
##sliding windows
import numpy as np
num_of_running_windows = 10
dfs_np=dfs.to_numpy()#转换为numpy是为了方便索引切片取数据并求均值
number_of_genes,num_of_samples = dfs_np.shape
print(f'{number_of_genes},{num_of_samples}')
poles = np.floor(np.linspace(0,num_of_samples-1,endpoint=True,num=num_of_running_windows+2))
'''

slicing：
    temp=[1,2,3,4,5,6,7]
    MATLAB：temp[1:3] ---> [1,2,3]
    Python: temp[1:3] --->[2,3] ！

'''
poles=poles.astype(np.int32).reshape(1,-1)
poles.shape
```

```python
window_start = poles[:,:-2].flatten()
window_end = poles[:,2:].flatten()
window_end[-1] = num_of_samples-1
window_end
```

```python

','.join(list(dfs.iloc[:,window_start[0]:window_end[0]+1].columns))
```

```python
np.mean([111.786406,0.000000,0.000000,0.000000,0.000000,0.000000,0.000000])
```

```python
#sliding
TPM_W=None
col_name4win=[]
for i in range(0,num_of_running_windows):
    wtpm_mean=np.mean(dfs_np[:,window_start[i]:window_end[i]+1],axis=1).reshape(-1,1)
    col_name4win.append(','.join(list(dfs.iloc[:,window_start[i]:window_end[i]+1].columns)))
    '''
        df1_np[:,window_start[i]:window_end[i]+1]
        np.mean(df1_np[:,window_start[i]:window_end[i]+1],axis=1)
    '''
    if TPM_W is None:
        TPM_W=wtpm_mean#numpy
    else:
        TPM_W=np.concatenate((TPM_W,wtpm_mean),axis=1)
print(TPM_W.shape)


TPM_W
```
