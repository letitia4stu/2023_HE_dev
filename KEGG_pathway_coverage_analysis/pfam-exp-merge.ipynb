```python
dfd = pd.read_excel('/Users/lichaoran/Downloads/figure2/pathplot/bv898slide-Notch-path-zhanshi-exp.xlsx', sheet_name='Sheet4')
dfd
```

```python
dfs1=pd.merge(dfc,dfd,on='Gene ID',how='inner')
dfs1
```

```python
dfs = dfs1.iloc[:, 0:30]
#dfs.drop_duplicates(inplace=True)
dfs.set_index('Gene ID',inplace=True)
dfs
```

```python
###huadong 窗口
#MATLAB与Python相比，索引与切片规则不同
#numpy与MATLAB函数也有不同
import numpy as np
num_of_running_windows = 10
dfs_np=dfs.to_numpy()#转换为numpy是为了方便索引切片取数据并求均值
number_of_genes,num_of_samples = dfs_np.shape
print(f'{number_of_genes},{num_of_samples}')
poles = np.floor(np.linspace(0,num_of_samples-1,endpoint=True,num=num_of_running_windows+2))
'''
MATLAB中，索引（下标）从1开始，Python中从0开始
切片：
    temp=[1,2,3,4,5,6,7]
    MATLAB中：temp[1:3] ---> [1,2,3]
    Python中: temp[1:3] --->[2,3] ！前含后不含

'''
poles=poles.astype(np.int32).reshape(1,-1)
poles.shape
```

```python
window_start = poles[:,:-2].flatten()
window_end = poles[:,2:].flatten()
window_end[-1] = num_of_samples-1#换成最后一个样本的索引
window_end
```

```python
#sliding
TPM_W=None
col_name4win=[]
for i in range(0,num_of_running_windows):
    wtpm_mean=np.mean(dfs_np[:,window_start[i]:window_end[i]+1],axis=1).reshape(-1,1)
    col_name4win.append(','.join(list(dfs.iloc[:,window_start[i]:window_end[i]+1].columns)))
    '''
        df1_np[:,window_start[i]:window_end[i]+1]
        np.mean(df1_np[:,window_start[i]:window_end[i]+1],axis=1)
    '''
    if TPM_W is None:
        TPM_W=wtpm_mean#numpy
    else:
        TPM_W=np.concatenate((TPM_W,wtpm_mean),axis=1)
print(TPM_W.shape)


TPM_W
```

```python
df2=pd.DataFrame(TPM_W,index=dfs.index,columns=['window_'+str(i) for i in range(1,11)])
df2=round(df2)
```

```python
dfhh=pd.merge(df2,dfd,on='Gene ID',how='inner')
dfhh
```
